```python
file = '../data/data.pkl'
with open(file, 'rb') as handle:
    data_dict = pickle.load(handle)
    
stocks_under_cosideration = ['AXISBANK', 'BAJFINANCE', 'BAJAJHLDNG', 'CHOLAFIN', 'HDFCBANK', 'ICICIBANK','KOTAKBANK', 'SBIN']

data_filtered = {}
for stocks in stocks_under_cosideration:
    data_filtered[stocks] = data_dict[stocks]
    
data_filtered['AXISBANK']
```

```python
for key in data_filtered.keys():
    data_filtered[key].reset_index(inplace = True, drop = True)
    data_filtered[key]['co_stkdate'] = pd.to_datetime(data_filtered[key]['co_stkdate'],format='%d-%m-%Y')
    data_filtered[key]['co_stkdate'] = data_filtered[key]['co_stkdate'].dt.strftime('%Y-%m-%d')
    #data_filtered[key]['daily_return'] = data_filtered[key]['nse_closing_price'].pct_change()
    data_filtered[key].set_index('co_stkdate', inplace = True)
```
